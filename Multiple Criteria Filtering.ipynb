# Applying multiple filter criter to a pandas DataFrame

```python
# import pandas library
import pandas as pd

emp = pd.read_csv("Datasets/employees.csv")
emp.head()
```

```python
emp[emp.Salary >= 100000]
```

### 2 conditions

* Salary > 100000
* Team only Finance

```python
# when you wrap conditions in parantheses, you give order
# you do those in brackets first before 'and'
# AND
emp[(emp.Salary > 140000) & (emp.Team == 'Finance')]
```

```python
# OR
emp[(emp.Salary > 140000) | (emp.Team == 'Finance')]
```

```python
(emp.Salary > 140000) | (emp.Team == 'Finance')
```

```python
(emp.Salary > 140000) & (emp.Team == 'Finance')
```

### What if you want Team Finance , Business Development and Marketing ?

```python
emp[(emp.Team == 'Finance') | (emp.Team == 'Business Development') | (emp.Team == 'Marketing')]
```

```python
# fast method
filter_list = ['Finance', 'Business Development', 'Marketing']
emp[emp.Team.isin(filter_list)]
```
